```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of plots we will create
```

```python
# Read in the data

df = pd.read_csv('movies.csv')
```

```python
# Let's Look at the data

df.head()
```

```python
# Let's see if there is any missing data
df = df.dropna()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.isnull().sum()
```

```python
# Data types for our columns

df.dtypes
```

```python
df.head()
```

```python
# change data type of columns
# df['budget'] = df['budget'].astype('int64')

# df['gross'] = df['gross'].astype('int64')

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 
```

```python
df.head()
```

```python
# Create correct year column
# pd.set_option('display.max_rows', 10 ) # displays no of rows you want to display
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
# df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Let's start looking at correlation
df.corr(method ='pearson', numeric_only=True) #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method ='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric features')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```
